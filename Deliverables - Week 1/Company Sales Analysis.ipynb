### Using Matplotlib  how do i carryout the following exercises   
Exercise 1: Read Total profit of all months and show it using a line plot.\
Exercise 2: Read Bathing soap facewash of all months and display it using the Subplot.

```python
#Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data  = pd.read_csv("company_sales_data.csv",header=0)
```

```python
print(data)
```

```python
#Exercise 1: Read Total profit of all months and show it using a line plot.
```

```python
# Line plot for total profit
plt.figure(figsize=(10, 5))
plt.plot(data.month_number, data.total_profit, marker='o', linestyle='-', color='b', label='Total Profit')
plt.title('Total Profit of All Months')
plt.xlabel('Months')
plt.ylabel('Profit in USD')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Exercise 2: Read Bathing soap facewash of all months and display it using the Subplot.
```

```python
plt.figure(figsize=(12, 6))

# Subplot 1: Bathing soap sales
plt.subplot(2, 1, 1)
plt.plot(data.month_number, data.bathingsoap, marker='o', linestyle='-', color='g', label='Bathing Soap Sales')
plt.title('Sales Data')
plt.ylabel('Bathing Soap Sales')
plt.legend()
plt.grid(True)

# Subplot 2: Facewash sales
plt.subplot(2, 1, 2)
plt.plot(data.month_number, data.facewash, marker='o', linestyle='-', color='r', label='Facewash Sales')
plt.xlabel('Months')
plt.ylabel('Facewash Sales')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
